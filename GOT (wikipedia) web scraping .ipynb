```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Game_of_Thrones_episodes"
```

```python
page = requests.get(url)
```

```python
page.status_code
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
episode_table = soup.find_all('table',class_='wikitable plainrowheaders wikiepisodetable')
print(episode_table)
```

```python
got=[]
for table in episode_table:
    rows = table.find_all('tr')[1:]
    for items in rows:
        season_number = items.find_all('td')[0].get_text()
        director = (items.find_all('td')[2].get_text())
        print(director)
        got_data = {'episode number': season_number,
                       'director':director
                   }
        got.append(got_data)
#         print(season_number, director)
        #df = pd.DataFrame(got)
```

```python
df = pd.DataFrame(season_number, director)
```

```python
episodes=[]

for table in episode_table:
    headers = []
    for header in table.find('tr').find_all('th'):
        headers.append(header.text)
    for row in table.find_all('tr')[1:]:
        value = []
        for col in row.find_all(['th', 'td']):
            value.append(col.text)
        episodes_dict = {headers[i]: value[i] for i in range(len(value))}
        episodes.append(episodes_dict)
```

```python
df = pd.DataFrame(episodes)
df
```

```python
df.drop(['Original air date' , 'No.'], axis=1,inplace = True)
```
